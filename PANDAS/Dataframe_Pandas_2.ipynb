```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python

df = pd.DataFrame(randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
df
```

Conditional Selection

```python
df > 0 # returns a boolean 
# if the data satisfies the condition then returns True
# else returns FALSE
```

```python
booldf = df>0
df[booldf] # adding to the original dataframe.
# returns all the values if the values are True 
# and returns NaN if the value is false
```

```python
df["W"]>0
```

```python
df[df['W']>0]
```

```python
df
```

```python
df[df['Z']>0]
```

```python
df[df['Z']>0][['X','Y']].loc['C']
```

```python
df[(df['W']>0) & (df['Y']>0)] # and condition, 
# true and true is true 
# true and false is false

# this returns the W and Y columns number 
# which is greater than 0 in all the rows 
```

```python
df[(df['W']>0) | (df['Y']>0)] # or condition
```
