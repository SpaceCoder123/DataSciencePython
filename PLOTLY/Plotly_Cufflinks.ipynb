# Plotly

Plotly is an interactive visualization library.
Cufflinks connect plotly with Pandas

```python
import pandas as pd
import numpy as np
from plotly import __version__
%matplotlib inline
```

```python
print(__version__) # not campatible if the version is less than 1.8
```

```python
import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,iplot
```

```python
init_notebook_mode(connected = True)
cf.go_offline()
```

```python
# data
df = pd.DataFrame(np.random.randn(100,4),columns = 'A B C D'.split())
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'values':[32,43,50]})
df2
```

```python
df.iplot()
```

```python
df.iplot(kind = 'scatter',x= 'A',y='B',mode = 'markers',size = 10)
```

```python
df2.iplot(kind = 'bar',x = 'Category')
```

```python
df.sum().iplot(kind = 'bar')
```

```python
df.iplot(kind = 'box')
# if you want to view certain plots 
# then click on the legend values to switch off the rest of the columns
```

3d Surface Plot

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,40,50],'z':[500,400,300,200,100]})
df3
```

```python
df3.iplot(kind = 'surface')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,40,50],'z':[5,4,3,2,1]})
df3.iplot(kind = 'surface',colorscale = 'rdylbu')
```

```python
df['A'].iplot(kind = 'hist',bins = 25) # histogram plot
```

```python
df.iplot(kind = 'hist',bins = 25) 
# histogram plot where values are overlapped on each other 
# and can be checked when the legend is switched off
```

```python
df[['A','B']].iplot(kind = 'spread') 
# first is just an iplot 
# second is the spread plot 
# where you can see the density or the points distribution of the plot
```
