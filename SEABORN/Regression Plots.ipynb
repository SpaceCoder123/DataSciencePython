# Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y ='tip',data = tips,hue = 'sex',markers = ['o','v'],scatter_kws = {'s':100})
# we get a scatter plot with a linear fit on it
# markers represent the data symbol for each category, if we see 
# all males points are in circles and females points are in nabla operator
```

```python
sns.lmplot(x='total_bill',y = 'tip',data = tips,col ='day',hue ='sex',aspect = 0.6,size =8)
```
