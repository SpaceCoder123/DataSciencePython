Seaborn is a statistical plotting library.
It has beautiful default plotting styles.
It has been designed to plot very well with pandas dataframes.

# Distribution plots 

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.head()
```

Dist plot, Allows us to visualize a univariable set of distributions or just one variable


```python
sns.distplot(tips['total_bill'],kde = False,bins = 70)
```

Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data = tips,kind='hex')
plt.tight_layout()
# hexa plot will just increase the hexagon color density if the points are more
sns.jointplot(x='total_bill',y='tip',data = tips,kind='reg')
sns.jointplot(x='total_bill',y='tip',data = tips,kind='kde')
#reg will return the regression line of the plot
```

```python
sns.pairplot(tips) 
# you can quickly visualize the plot data
# creates all plots with numerical values and if all the values are common 
#the it makes a box plot
sns.pairplot(tips,hue = 'sex')
# if any plot is categorical then it plots with respect to those plots,
# In the below example, 
#you can see all male plots are in blue and female are orange
sns.pairplot(tips,hue = 'sex',palette ="cool") # changed the color of the data
```

```python
sns.rugplot(tips['total_bill'])
```
