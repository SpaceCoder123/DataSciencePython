# Categorical Plots

```python
import seaborn as sns # imports seaborn
%matplotlib inline 
tips = sns.load_dataset('tips') # loads the default dataset which is present
tips.head()
```

```python
import numpy as np
```

# Bar plot

```python
sns.barplot(x ='sex',y ='total_bill', data = tips,estimator =np.std ) 
# averaged totol bill for males and females altogether
#estimator is a function on which the data will be sorted # 
# and in the current data the value is according to the standard deviation
# estimator is explicity calculate the 
# standard deviation of the total bill with respect to male or female
```

```python
sns.countplot(x='sex',data = tips)
# the plot will onlt count the accurances of the tips with respect to male or female
# result, the number of men are more than women
```

```python
sns.boxplot(x='day',y='total_bill',data = tips,hue ='smoker')
# this plot shows the total bill distribution on the given data set 
# and categorically divides the set of values based on 
# whether the person is a smoker or not
# very simple and easy to understand
```

```python
sns.violinplot(x='day',y='total_bill',data = tips,hue = 'sex',split = True)
# this plot actually shows all the components corresponding to the given dataset
# it actually shows all the points corresponding to the current dataset
# this plot is currently showing the kernel density estimation of the underlying distribution
# the split divides the categorical datatypes of male and female but merges it to one voilin plot
# now this plot returns the most number of points in the given area
```

```python
sns.stripplot(x = 'day',y ='total_bill',data= tips,jitter = True,hue = 'sex',split = True)
# this is just a scatter plot based on the category
# one drawback is of this plot is that
# you cant tell how many points are stacked upon each other
# to counter this problem, we use jitter is equal to true
```

```python
sns.swarmplot(x ='day',y ='total_bill',data = tips)
# this is combination of stripplot and violinplot
# more computational power is required 
# to place the elements next to each other and therefore not very useful
# not regularly used and not recommended
```

```python
sns.violinplot(x ='day',y ='total_bill',data = tips)
sns.swarmplot(x ='day',y ='total_bill',data = tips,color = 'black')
```

```python
sns.factorplot(x='day',y ='total_bill',data = tips,kind = 'violin') 
# more of a generic plot and u can call any type of plotting using kind method
```
