```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure(figsize = (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,color = "purple",label = 'X squared',linewidth = "3",alpha = 0.5) # increases the width of the line
ax.plot(x,x**3,label = "X cubed",linestyle = '--',marker = 'o',markersize = 20,markerfacecolor ='yellow')
ax.set_title("Title")
ax.set_ylabel("Y")
ax.set_xlabel("X")
ax.legend(loc = 10) # keeps the location at a particular place
# alpha command talks about the transparency of the plot
# marker command shows the points that are present in the plot

#set limits, the lower bounds and upper bounds
ax.set_xlim([0,1])
ax.set_ylim([0,2]) # these commands will cap the data upto a defined limit
```
