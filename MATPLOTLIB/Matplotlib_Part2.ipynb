```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 2)
axes.plot(x,y)
plt.tight_layout() # fixed the overlapping issue
```

# Figure Size and DPI

```python
fig = plt.figure(figsize = (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X squared')
ax.plot(x,x**3,label = "X cubed")
ax.set_title("Title")
ax.set_ylabel("Y")
ax.set_xlabel("X")
ax.legend(loc = 10) # keeps the location at a particular place
fig.savefig('my_pix1.jpeg',dpi = 300)
```

```python
fig, axes = plt.subplots(nrows = 2,ncols = 1,figsize =(4,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
